```python
import datetime

def find_after_min(now ,after_min):
    #csv 자료는 2분 간격으로 한 행이 기록되어 있다.
    #after_min 만큼 뒤의 실제 데이터를 찾아서 반환한다

    #### 입력 #### 
    # now = '2022-12-15 0:00' 와 같은 형식의 문자열
    # after_min = 지난 시간을 분 단위로 나타낸 숫자

    
    date, time= now.split(' ')

    year, month, day = date.split('-')
    hour, minit = time.split(':')

    year = int(year)
    month = int(month)
    day = int(day)
    hour = int(hour)
    minit = int(minit)

    #시간 얻어오기
    after_min = hour * 60 + minit + after_min
    date = datetime.datetime(year, month, day) + datetime.timedelta(minutes = after_min)


    #데이터프레임에서 조회하기
    d1, d2, d3 =str(date).split(':') 
    dat, dh = d1.split(' ')
    if (int(dh) < 10):
        dh = ' ' +  dh[1]
        d1 = dat+dh

    d_all = d1 + ':' +d2
    #print(d_all)
    after_data = merge_outer_Null[merge_outer_Null['일시']==d_all]
    print(after_data)
    
    return after_data
```

```python
## test

d = find_after_min('2022-12-25 00:20' ,30)
print(d)

d = find_after_min('2022-12-25 01:20' ,30)
print(d)

d = find_after_min('2022-12-25 1:2' ,30)
print(d)
```
