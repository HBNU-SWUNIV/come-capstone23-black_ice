##자동화!

```python
def preprocessing(csv_path, save_path):
    surface_temp_csv = pd.read_csv(csv_path)
    name = csv_path.split('/')[-1].split('.')[0] + '_p'

    st_drop = surface_temp_csv[ ['Time', '표면'] ].dropna(subset=['표면'], ignore_index=True)
    if save_path[-1] != '/':
        save_path = save_path + '/'
    st_drop.to_csv( save_path + name + '.csv' )
```

```python
cp = '../model/Data_surface/surface_23_02_off.csv'
sp =  '../model/Data_pre_processing'
preprocessing(cp, sp)
```
