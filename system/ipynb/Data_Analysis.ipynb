```python
import math
##### 이슬점 근사치를 계산한다. 

## 상수 ## 
b = 17.62
c = 243.12

t = -10 #온도
rh = 76 #습도

gamma = (b * t/(c+t)) + math.log(rh/100.0)
dewpoint = (c * gamma) / (b - gamma)

print('이슬점', dewpoint )  # 이슬점 온도는 습도가 100%가 되면 현재 온도와 같아진다. (보통은 현재 온도보다 낮다!)
# -9.683906746496984 (-5, 69.7) 
# -14.38465776211111 (-7.1, 56.1)
```
